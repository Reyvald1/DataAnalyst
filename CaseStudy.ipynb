```python
# Replace 'Release Date of The Game' Object Data Type to datetime Data Type

tabel_clear = tabel.copy()
tabel_clear['Release'] = pd.to_datetime(tabel_clear['Release'])

print(tabel_clear.dtypes)
tabel_clear.head()
```

```python
# check typo

tabel_clear['Name'].value_counts()
```

```python
#check is there any duplicated name

tabel_clear['Name'].duplicated().sum()
```

```python


tabel_clear.drop_duplicates(inplace = True)
print(tabel_clear.to_string())
```

```python
#re-check is there any duplicated name

tabel_clear['Name'].duplicated().sum()
```

```python
# check typo

for col_name in ['Genre', 'Developer', 'Publisher',]:
       print(tabel_clear[col_name].value_counts().to_string(),'\n')
```

```python
# Classify Nan Series as Non-Series
tabel_clear[ 'Series' ] = tabel_clear[ 'Series' ].replace(np.nan, 'Non-Series')
print(tabel_clear)
```

# PUBLISHER WHO PUBLISHED MOST OF THE GAME



```python
# Add a Column to Sum the Object
tabel_clear = tabel_clear.assign(Quantity = 1)
tabel_clear.head()
```

```python
most_pub = tabel_clear.groupby('Publisher', as_index=False)['Quantity'].sum()
most_pub2 = most_pub.sort_values(by=['Quantity'], ascending=False)
most_pub2.head()
```

# DEVELOPER WHO DEVELOP MOST OF THE GAME

```python
most_dev = tabel_clear.groupby('Developer', as_index=False)['Quantity'].sum()
most_dev2 = most_dev.sort_values(by=['Quantity'], ascending=False)
most_dev2.head()
```

# SERIES THE MOST SALES

```python
most_dev = tabel_clear.groupby('Sales', as_index=False)['Series'].sum()
most_dev2 = most_dev.sort_values(by=['Sales'], ascending=False)
most_dev2.head()
```

# SERIES THE MOST GAME

```python
most_dev = tabel_clear.groupby('Series', as_index=False)['Quantity'].sum()
most_dev2 = most_dev.sort_values(by=['Quantity'], ascending=False)
most_dev2.head()
```
